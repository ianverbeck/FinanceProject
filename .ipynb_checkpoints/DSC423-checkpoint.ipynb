```python
import yfinance as yf
import yahoo_fin.stock_info as si
#import yahoofinancials

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')

import datetime as dt
from dateutil.relativedelta import relativedelta
from timeit import default_timer as timer

import pickle
```

```python
with open('dowticker2021-02-04.pickle', 'rb') as handle:
    historical_dow = pickle.load(handle)
```

# 1.2

```python
ticker = 'AAPL'
```

```python
print('========'+ticker+'=======')

display(historical_dow[ticker].close.describe())

fig, axes = plt.subplots(1,1,figsize=(15,8))
historical_dow[ticker].close.plot(kind='line')
plt.show()
```

```python
dow_fundamentals = pd.DataFrame(columns = si.get_quote_table(ticker).keys())
for ticker in historical_dow:
    qt = si.get_quote_table(ticker)
    append = pd.Series([val for val in qt.values()],index = qt.keys(),name=ticker)
    dow_fundamentals = dow_fundamentals.append(append)
```

```python
dow_fundamentals.isnull().sum(axis=0)
```

```python
dow_fundamentals['PE Ratio (TTM)'].plot(kind='hist',bins=len(historical_dow));
```
